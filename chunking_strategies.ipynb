```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)

import PyPDF2
```

```python
import PyPDF2

pdf_path = "data\hleg_ethics_guidelines_for_trustworthy_ai-en.pdf"

pdf_text = ""

with open(pdf_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)

    for page in reader.pages:
        text = page.extract_text()
        if text:
            pdf_text += text + "\n"

print(f"PDF loaded: {len(pdf_text)} characters")
print(pdf_text[:500])
```

```python
print(f"Total PDF characters: {len(pdf_text):,}")

pdf_words = len(pdf_text.split())

print(f"Approximate words: {pdf_words:,}")
```

```python
start = 22000
end = 27000

pdf_sample = pdf_text[start:end]

print(pdf_sample)
```

```python
fixed_splitter = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100,
    separator="\n"
)

pdf_chunks_fixed = fixed_splitter.split_text(pdf_text)

print(f"Number of chunks: {len(pdf_chunks_fixed)}")

for i in range(3):
    print(f"\n--- Chunk {i+1} ---\n")
    print(pdf_chunks_fixed[i][:300])
```

### Fixed-size PDF chunking observations

Using CharacterTextSplitter with chunk_size=1000 and chunk_overlap=100 created 171 chunks. The chunks are mostly readable, but several boundaries cut sentences or context awkwardly. This shows that fixed-size chunking is simple and predictable, but it does not reliably preserve sentence or section boundaries in the PDF.

```python
recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len,
    separators=["\n\n", "\n", ". ", " ", ""]
)

pdf_chunks_recursive = recursive_splitter.split_text(pdf_text)

print(f"Number of chunks: {len(pdf_chunks_recursive)}")

for i in range(3):
    print(f"\n--- Chunk {i+1} ---\n")
    print(pdf_chunks_recursive[i][:300])
```

```python
print(pdf_chunks_fixed[50][:500])

print("\n" + "="*50 + "\n")

print(pdf_chunks_recursive[50][:500])
```

### Recursive PDF chunking observations

RecursiveCharacterTextSplitter created 196 chunks. Compared with fixed-size chunking, the chunks were more likely to start at meaningful section or paragraph boundaries, such as section headings. However, some chunks were still quite short or incomplete because the extracted PDF text contains formatting artefacts and irregular line breaks. Overall, recursive chunking preserved document structure better than fixed-size chunking, but the result still depends heavily on the quality of PDF text extraction.

```python
token_splitter = TokenTextSplitter(
    chunk_size=250,
    chunk_overlap=50
)

pdf_chunks_token = token_splitter.split_text(pdf_text)

print(f"Number of chunks: {len(pdf_chunks_token)}")

for i in range(3):
    print(f"\n--- Chunk {i+1} ---\n")
    print(pdf_chunks_token[i][:300])
```

```python
print(pdf_chunks_token[50][:500])
```

### Token-based PDF chunking observations

Token-based chunking produced chunks of a more predictable size for use with large language models. However, it did not consistently preserve document structure or section boundaries. Compared with recursive chunking, the resulting chunks appeared less semantically coherent, although they remained more suitable for managing LLM context limits than fixed-size character chunking.
